```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
eps = 0.0001
q = 0.9

def g(x, r):
    return r * x * (1 - x)

def doSimpleIterations(x0, r):
    counter = 0
    x = g(x0, r)
    c = abs(x - x0) * q / (1 - q)
    x_arr = []
    x_arr.append(x0)
    x_arr.append(x)
    while c > eps:
        next_x = g(x, r)
        c = abs(next_x - x) * q / (1 - q)
        x = next_x
        x_arr.append(x)
        counter = counter + 1
        if (counter > 2000):
            return x_arr
    return x_arr
```

```python
y_diag = []
x_diag = []
r_c = 0.01
for xx in range(1, 50):
    x_cu = xx * 0.02
    r_c = 2.0
    for j in range(2000, 4000):
        y = doSimpleIterations(x_cu, r_c)
        if (y[-1] < 0):
            continue
        if (r_c > 1 and y[-1] == 0): 
            continue
        y_diag.append(y[-1])
        x_diag.append(r_c)
        r_c += 0.001
fig=plt.figure(figsize = (15, 10))
axes=fig.add_subplot(111)
axes.plot(x_diag,y_diag, 'g.', markersize = 1)
plt.show()
```

```python
r_last = 0.5
x_last = 0.2
def build_path(xx, rr, y_g, x_g, zero_arr):
    x_arr = doSimpleIterations(xx, rr)
    ys = []
    xs = []
    for i in range (0, len(x_arr) - 1):
        x = x_arr[i]
        xs.append(x)
        ys.append(x)
        xs.append(x)
        y = g(x, rr)
        ys.append(y)
    xs[0] = xx
    ys[0] = 0
    return xs, ys

def draw_g(xx, rr):
    y_g = []
    x_g = []
    zero_arr = []
    for x_inter in range(0, 100, 1):
        x = x_inter * 0.01 
        y = g(x, rr)
        x_g.append(x)
        y_g.append(y)
        zero_arr.append(0)
    xs, ys = build_path(xx, rr, x_g, y_g, zero_arr)
    fig= plt.figure(figsize = (15, 10))
    axes=fig.add_subplot(111)
    axes.plot(x_g, y_g)
    axes.plot(x_g, x_g)
    axes.plot(x_g, zero_arr)
    axes.plot(xs, ys)
    plt.show()
draw_g(x_last, r_last)
```
